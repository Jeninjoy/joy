```python
def hah(age):
    if age == '5-14 years':
        return 0
    elif age=='15-24 years':
        return 1
    elif age=='25-34 years':
        return 2
    elif age=='35-54 years':
        return 3
    elif age=='55-74 years':
        return 4
    else:
        return 5
data['age']=data['age'].apply(hah)    
```

```python
data.head()
```

```python
data['country']=pd.factorize(data['country'])[0]
data['year']=pd.factorize(data['year'])[0]
data['sex']=pd.factorize(data['sex'])[0]
```

```python
x=data[['country','year','sex','age','population']]
y=data['suicides_no']
```

```python
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
lm.fit(x,y)
```

```python
y_pred=lm.predict(x)
```

```python
import statsmodels.formula.api as sm
model_accuracy=sm.OLS(y,x).fit()
model_accuracy.summary()
```
